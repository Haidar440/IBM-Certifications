## Let's learn about how to combine python and excel
We'll be primarily using two new libraries that are;
1. openpyexcel (package name: pyexcel-xlsx): Helps use to load post 2009 excel files and perform analysis on them.
- > Features:
  > 1. Read/write Excel files.
  > 2. Create new workbooks, modify cells, styles, charts.
  > 3. Supports formulas and formatting.
**Installation:** `pip install openpyxl` [*add a `!` at the begining if you are installing it inside a notebook*]
- Documentation: [Link](https://openpyxl.readthedocs.io/en/stable/)

2. xlsxwriter:
- > Features:
  > 1. Write Excel files with advanced formatting.
  > 2. Supports charts, conditional formatting, images.
  > 3. Write-only (cannot read files).
**Installation:**: `pip install xlsxwriter`[*add a `!` at the begining if you are installing it inside a notebook*]
- Documentation: [Link](https://xlsxwriter.readthedocs.io/)
